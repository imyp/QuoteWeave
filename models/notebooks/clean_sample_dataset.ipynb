```python
import pandas as pd
```

```python
df = pd.read_csv("hf://datasets/jstet/quotes-500k/quotes.csv")
```

```python
df
```

```python
def clean_dataset(df):
    # Clean 'quote' column
    df = df[df["quote"].notna()]
    df["quote"] = df["quote"].astype(str).str.strip()
    df = df[df["quote"] != ""]

    # Clean 'author' column
    df = df[df["author"].notna()]
    df["author"] = df["author"].astype(str).str.strip()
    df = df[df["author"] != ""]

    # Clean 'category' column
    df["category"] = (
        df["category"].fillna("").astype(str)
    )  # Fill NaNs and ensure string type
    df["category"] = df["category"].apply(
        lambda x: [cat.strip().lower() for cat in x.split(",") if cat.strip()]
    )
    # Optional: Remove rows where category list is empty after cleaning, if desired
    df = df[df["category"].apply(lambda x: len(x) > 0)]

    return df
```

```python
clean_dataset(df)[:500].to_csv("quotes_sample.csv", index=False)
```
